```python
import numpy as np # For Array/Matrix Manipulattttion
import pandas as pd # For Reading CSV FILE FROM Google_Collab
import matplotlib.pyplot as plt # Sketching the graph

class NeuralNetwork():
    def __init__(self, input_size, hidden_nodes, output_size, learning_rate=0.1):
        """
        :param input_size: Number of input neurons (784 for MNIST)
        :param hidden_nodes: List specifying number of neurons in each hidden layer
        :param output_size: Number of output neurons (10 for MNIST digits 0-9)
        :param learning_rate: Learning rate for weight updates
        """
        self.input_size = input_size
        self.hidden_nodes = hidden_nodes
        self.output_size = output_size
        self.learning_rate = learning_rate

        # Define the Model
        layer_sizes = [input_size] + hidden_nodes + [output_size]

        # Initialize weights and biases
        self.weights = [np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.01 for i in range(len(layer_sizes) - 1)]
        self.biases = [np.zeros((1, layer_sizes[i+1])) for i in range(len(layer_sizes) - 1)]

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def feedForward(self, inputs):
        self.layers = [inputs]  # Store outputs of the layers in this
        for i in range(len(self.weights) - 1):
            inputs = self.sigmoid(np.dot(inputs, self.weights[i]) + self.biases[i])
            self.layers.append(inputs)

        # Output layer uses softmax for Probability Mapping...
        outputs = self.softmax(np.dot(inputs, self.weights[-1]) + self.biases[-1])
        self.layers.append(outputs)
        return outputs

    def cross_entropy_loss(self, predictions, targets):
        return -np.sum(targets * np.log(predictions + 1e-9)) / len(targets) # Normalise to avoid log(0) -> Infinity

    def backpropagation(self, target_output):
        errors = [self.layers[-1] - target_output]  # Output layer error
        deltas = [errors[0]]

        # Backprop back to the hidden layers
        for i in range(len(self.hidden_nodes), 0, -1):
            errors.insert(0, np.dot(deltas[0], self.weights[i].T))
            deltas.insert(0, errors[0] * self.sigmoid_derivative(self.layers[i]))

        # Update Weigts and biases
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * np.dot(self.layers[i].T, deltas[i])
            self.biases[i] -= self.learning_rate * np.sum(deltas[i], axis=0, keepdims=True)

    def train(self, X, y, epochs=10, batch_size=64):
        for epoch in range(epochs):
            total_loss = 0
            for i in range(0, len(X), batch_size):
                batch_X = X[i:i+batch_size]
                batch_y = y[i:i+batch_size]

                predictions = self.feedForward(batch_X)
                loss = self.cross_entropy_loss(predictions, batch_y)
                total_loss += loss

                self.backpropagation(batch_y)

            print(f"Epoch {epoch+1}, Loss: {total_loss / (len(X) / batch_size):.4f}")

    def predict(self, X):
      probabilities = self.feedForward(X)
      predicted_labels = np.argmax(probabilities, axis=1)  # Get the predicted
      confidence_scores = np.max(probabilities, axis=1)  # Get confidence

      return [predicted_labels, confidence_scores]


    def Prepare_Data(self, train_data, test_data):
      #  Prep The Data
      X_train = train_data.iloc[:, 1:].values / 255.0  # Normalize pixel vals, To be in range 0-1
      y_train = one_hot_encode(train_data.iloc[:, 0].values)

      # Prep TEST data
      X_test = test_data.iloc[:, 1:].values / 255.0  # Normalize
      y_test = test_data.iloc[:, 0].values

      return X_train, y_train, X_test, y_test

    def Test(self, X_test, y_test, index):
      sample_image = X_test[index].reshape(28, 28)  # reshape to 28x28 for display
      true_label = y_test[index]  # Get the real Label

      # Display the image
      plt.imshow(sample_image, cmap='gray')
      plt.title(f"Actual Label: {true_label}")
      plt.axis('off')
      plt.show()

      # Predict using the neural network
      predicted_label, confidence = nn.predict(X_test[index].reshape(1, -1))
      print(f"Predicted Label: {predicted_label[0]}, Confidence : {confidence[0]*100:.2f}%")


    def save_model(self, filename="model"):
        """Save model weights and biases to a compressed file."""
        # Save weights and biases separately to avoid inhomogeneous shape error
        # Convert weights and biases to lists before saving
        np.savez_compressed(filename, weights=np.array([w.tolist() for w in self.weights], dtype=object), biases=np.array([b.tolist() for b in self.biases], dtype=object))
        print(f"Model saved as {filename}.npz")

    def load_model(self, filename="model"):
        """Load model weights and biases from a compressed file."""
        data = np.load(filename + ".npz", allow_pickle=True)
        self.weights = list(data["weights"])
        self.biases = list(data["biases"])
        print(f"Model loaded from {filename}.npz")


def one_hot_encode(labels, num_classes=10):
    return np.eye(num_classes)[labels]


# Load MNIST dataset from Google Colab sample_dataFolder
train_data = pd.read_csv('/content/sample_data/mnist_train_small.csv', header=None)
test_data = pd.read_csv('/content/sample_data/mnist_test.csv', header=None)


# Create Neural Network: 784 input(28*28 pixels), 1 hidden layer with 128 neurons, 10 output[0-9]
nn = NeuralNetwork(input_size=784, hidden_nodes=[128], output_size=10, learning_rate=0.01)

# Prep Data
X_Train, Y_Train, X_Test, Y_Test = nn.Prepare_Data(train_data, test_data=test_data)
# Train the nn On the prepared, pre-processed, Cleaned Data
nn.train(X_Train, Y_Train, epochs=50, batch_size=64)

# Test the model
predictions, _ = nn.predict(X_Test)
accuracy = np.mean(predictions == Y_Test) * 100
print(f"Test Accuracy: {np.round(accuracy)}%")
```
